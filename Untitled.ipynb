```python
import pandas as pd
import matplotlib.pyplot as plt

#create data set

df = pd.DataFrame ({'hours': [1, 2, 4, 5, 6, 6, 7, 8, 10, 11, 11, 12, 12, 14],
                    'score': [64, 66, 76, 73, 81, 83, 82, 80, 88, 84, 82, 91, 93, 89]})

df[0:5]
df[0:7]

plt.scatter (df.hours, df.score)
plt.title ('Hours vs score')
plt.xlabel ('h')
plt.ylabel ('score')
plt.show()
```
